```python
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
#y = pd.factorize(df['price'])[0].reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

Not fitting test set

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.fit(X_train, y_train,validation_data = (X_test,y_test),batch_size = 128, epochs = 500)
```

Smaller the batch size, longer the training

Batch size normally in power of 2

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Blue - training loss,
Orange - Validation loss

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
mean_squared_error( y_test,pred)
```

```python
np.sqrt(mean_squared_error(pred, y_test))
```

```python
mean_absolute_error(y_test, pred)
```
